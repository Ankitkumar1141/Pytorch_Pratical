```python
import torch
print(torch.__version__)
```

## Tensors

```python
## using empty
torch.empty(4,4)
```

It allocates memory to tensors and shows values present at that memory.

```python
# using zero
torch.zeros(2,2)
```

It create tensor with all values as 0.

```python
# using ones 
torch.ones(2,2)
```

It create tensors with all values 1.

```python
torch.rand(2,3)
```

It create tensors with all random values between 0 to 1. Limitation is after every run of the same code will produce different tensors value.

```python
## To overcome above limitation
torch.manual_seed(100)
torch.rand(2,3)
```

```python
## using tensor
torch.tensor([[5,10,15],[20,25,30]])
```

It is used to generate customize or user defined tensors where user has to pass all values to be stored in tensors.

## Tensor shape

```python
t1 = torch.tensor([[1,2],[3,4],[4,5],[6,7]])
```

```python
t1.shape
```

## Tensor data type

```python
t1.dtype
```

```python
## declare data type of tensor at the time of executation
t3 = torch.tensor([[1,2],[3,5]],dtype=torch.float32)
```

# Mathematical operations

```python
## Scalar addition
print(t1 + 10)
```

All operations are same except operators.

```python
print(t1-1)
print(t1*4)
print(t1**2)
print(t1/2)
```

```python
## Elements wise operations
```

```python
t2 = torch.tensor([[1,2],
                   [2,3],
                   [3,4],
                   [4,5]])
```

```python
t1+t2
```

```python
t1-t2
t1*t2
t1/t2
```

```python
## Reduction operation
```

Reduction operation are used to reduce the tensor to a scalar value.

```python
print(torch.sum(t2))
```

```python
print(torch.mean(t3))
```

```python
print(torch.median(t3))
```

```python
print(torch.std(t3))
```
