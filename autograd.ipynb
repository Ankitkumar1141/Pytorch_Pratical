## Autograd
Autograd is an automated component present in Pytorch used to compute gradients.It is highly used in training deep neural networks(backpropagation).

```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad = True)
```

requires_grad :  It is parameter used to mention any tensor will be used for gradient computation, by default it is False.

```python
y = x**2
```

```python
z = torch.sin(y)
```

```python
print(y)
print(z)
```

```python
z.backward()
```

```python
x.grad
```
